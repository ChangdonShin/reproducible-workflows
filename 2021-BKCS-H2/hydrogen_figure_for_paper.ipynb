# 1. Import packages

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import matplotlib.colors as mlabcolor
import matplotlib.style as style
style.use('seaborn-poster')
import seaborn as sns
sns.set_palette("Paired",8)
```

# 2. Fit compressibility factor data to rational function 

```python

df0 = pd.read_csv("./com_exp.csv")

def com_fit_exp(P, e, f, g, h):
    Z=((1+e*P+f*P**2)/(1+g*P+h*P**2))
    return Z

pars, pcov = curve_fit(com_fit_exp, df0["P"], df0["Z"], [1,1,1,1])

print(pars)
print(pcov)

parameter_com_fit = pars

fig=plt. figure(figsize =(13,10)) 
ax0=fig.add_subplot(1,1,1)

plt.scatter(df0["P"], df0["Z"], s=100, color='blue', label='data')

qfit_com_exp = np.linspace(0,20,100)

plt. xlim(0,20)
plt. ylim(0.8,1.3)
plt. xticks(fontsize=28)
plt. tick_params(axis='both', width=4, length=10)
plt. yticks(fontsize=28)
for axis in ['top','bottom','left','right']:
    ax0.spines[axis].set_linewidth(4)

plt.plot(qfit_com_exp, com_fit_exp(qfit_com_exp,*parameter_com_fit), lw=4.5, markersize=10, color='black', label='fit')
plt.xlabel('Pressure / MPa', fontsize=36)
plt.ylabel('Compressibility Factor / -', fontsize=36)
plt.legend(frameon=True, loc = 'best',fontsize=28)
plt.tight_layout()
plt.savefig("./com_exp_fitting_results_0630.png",format='png')
plt.show()

fit_compare_com_exp = []

for k in df0["P"]:
    
    fit_compare_com_exp_value = com_fit_exp(k, *parameter_com_fit)
    fit_compare_com_exp.append(fit_compare_com_exp_value)

ss_res = np.dot(df0["Z"]-fit_compare_com_exp,df0["Z"]-fit_compare_com_exp)
mean = np.mean(df0["Z"])
ss_tot = np.dot(df0["Z"]-mean,df0["Z"]-mean)
r_squared_com_exp = 1-ss_res/ss_tot

print("R2 ="+str(r_squared_com_exp))
```

# 3. 3D-MOF case (IRMOF series)

## 3-1. IRMOF-1 (PLD: 7.9)

### 3-1-1. Fit data to function

```python

df0 = pd.read_csv("./com_exp.csv")

def com_fit_exp(P, e, f, g, h):
    Z=((1+e*P+f*P**2)/(1+g*P+h*P**2))
    return Z

pars, pcov = curve_fit(com_fit_exp, df0["P"], df0["Z"], [1,1,1,1])

parameter_com_fit = pars

df1 = pd.read_csv("./IRMOF_1.csv")

def excess_fit(P, b, c, V, D):
    m=(D-(1/com_fit_exp(P,*parameter_com_fit))*(P/(0.004157*77)))*0.001*V*((b*P)/((1+(b*P)**c)**(1/c)))*100
    return m

pars, pcov = curve_fit(excess_fit, df1["P"], df1["wt_e"]/10, [10000,0.3,0.8,50])

print(pars)
print(pcov)

parameter_IRMOF_1 = pars

fig=plt. figure(figsize =(13,10)) 
ax1=fig.add_subplot(1,1,1)

plt.scatter(df1["P"], df1["wt_e"]/10, s=100, color='blue', label='simulated excess isotherm')
plt.scatter(df1["P"], df1["wt_a"]/10, s=100, color='green', label='simulated absolute isotherm')

qfit_IRMOF_1 = np.linspace(0,10,100)

plt. xlim(0,10)
plt. ylim(0,10)
plt. xticks(fontsize=36)
plt. tick_params(axis='both', width=4, length=10)
plt. yticks(fontsize=36)
for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(4)

plt.plot(qfit_IRMOF_1, excess_fit(qfit_IRMOF_1,*parameter_IRMOF_1), lw=4.5, markersize=10, color='black', label='model fitting')
plt.xlabel('Pressure / MPa', fontsize=44)
plt.ylabel('H$_{2}$ Uptake / wt.%', fontsize=44)
plt.legend(frameon=True, loc = 'best',fontsize=32)
plt.tight_layout()
plt.savefig("./fitting_results_IRMOF_1.png",format='png')
plt.show()

fit_compare_IRMOF_1 = []

for i in df1["P"]:
    
    fit_compare_value_IRMOF_1 = excess_fit(i, *parameter_IRMOF_1)
    fit_compare_IRMOF_1.append(fit_compare_value_IRMOF_1)

ss_res = np.dot(df1["wt_e"]/10-fit_compare_IRMOF_1,df1["wt_e"]/10-fit_compare_IRMOF_1)
mean = np.mean(df1["wt_e"]/10)    
ss_tot = np.dot(df1["wt_e"]/10-mean,df1["wt_e"]/10-mean)
r_squared_IRMOF_1 = 1 - ss_res/ss_tot

print("R2 ="+str(r_squared_IRMOF_1))
```

### 3-1-2. Sensitivity analysis

```python

R_list=[]
D_list=[]
V_list=[]
c_obtained = parameter_IRMOF_1[1]
b_obtained = parameter_IRMOF_1[0]

print(b_obtained,c_obtained)

fig=plt. figure(figsize =(14,10)) 
ax11=fig.add_subplot(1,1,1)

#plt.xscale('log')
plt.xlim(0.0,4.0)
plt.ylim(40,90)
plt.xticks(fontsize=36)
plt.tick_params(axis='both', width=4, length=10)
plt. xticks(np.arange(0,5,1))
plt.yticks(fontsize=36)
for axis in ['top','bottom','left','right']:
    ax11.spines[axis].set_linewidth(4)
    
for d_test in np.linspace(40,90,100):
    for v_test in np.linspace(0.0,4.0,100):
                
        sensitivity_IRMOF_1 = []
                
        for j in df1["P"]:
            
            sensitivity_IRMOF_1_value = excess_fit(j, b_obtained, c_obtained, v_test, d_test)
            sensitivity_IRMOF_1.append(sensitivity_IRMOF_1_value)
        
        ss_res = np.dot(df1["wt_e"]/10-sensitivity_IRMOF_1,df1["wt_e"]/10-sensitivity_IRMOF_1)
        mean = np.mean(df1["wt_e"]/10)
        ss_tot = np.dot(df1["wt_e"]/10-mean,df1["wt_e"]/10-mean)
        r_squared_sen_IRMOF_1 = 1-ss_res/ss_tot
        R_list.append(r_squared_sen_IRMOF_1)
        D_list.append(d_test)
        V_list.append(v_test)

        
print("finish")

scat = ax11.scatter(V_list, D_list, c=R_list, s=100, cmap='Blues', norm=mlabcolor.Normalize(vmax=1.000,vmin=0.95))
# cmap = viridis, Blues, gray, autumn
color=fig.colorbar(scat, ticks=np.arange(0.95,1.01,0.01))
color.ax.set_title('R$^{2}$', fontsize=44, FontWeight='bold')
color.ax.tick_params(width=4, length=8, labelsize=36)
color.outline.set_linewidth(4)

x_range = np.linspace(0.0,4.0,100)
true_density = np.linspace(65.48736704,65.48736704,100)

y_range = np.linspace(40,90,100)
true_volume = np.linspace(1.41002026,1.41002026,100)

plt.scatter(parameter_IRMOF_1[2], parameter_IRMOF_1[3], s= 450, color='red', edgecolors='black', linewidths=2)

plt.plot(x_range, true_density, lw=4.5, linestyle='-.',color='black')
plt.plot(true_volume, y_range, lw=4.5, linestyle='--',color='black')

plt.xlabel('V$_{\mathrm{P}}$ / cm$^{3}$ g$^{{\u2212}1}$', fontsize=44)
plt.ylabel('ρ$_{\mathrm{A}}$ / kg m$^{{\u2212}3}$', fontsize=44)
plt.tight_layout()
plt. savefig("./parameter_sensitivity_IRMOF_1_D_vs_V.png",format='png')
plt.show()
```
