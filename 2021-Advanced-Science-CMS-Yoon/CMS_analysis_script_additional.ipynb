## 1-1. Ar saturation loading data from the  simulation

```python
import numpy as np
import pulp as p
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from scipy.optimize import fmin_cobyla

d_three_range_1 = [3.9, 4.4, 4.9, 5.4]
d_three_range_2 = [5.4, 5.6, 5.8]
d_three_range_3 = [5.8, 6.3, 6.8, 7.3]

l_three_range_1 = [128.360331, 133.2972668, 136.0742932, 158.875198]
l_three_range_2 = [158.875198, 197.2719015, 224.2206285]
l_three_range_3 = [224.2206285, 238.8242697, 252.7094017, 257.3670119]

def linear_Ar_SL(x, a, b):
    y=a*x+b
    return y

pars, pcov = curve_fit(linear_Ar_SL, d_three_range_1, l_three_range_1, [1,1])

parameter_three_Ar_SL_range_1 = pars 

pars, pcov = curve_fit(linear_Ar_SL, d_three_range_2, l_three_range_2, [1,1])

parameter_three_Ar_SL_range_2 = pars 

pars, pcov = curve_fit(linear_Ar_SL, d_three_range_3, l_three_range_3, [1,1])

parameter_three_Ar_SL_range_3 = pars 

print(parameter_three_Ar_SL_range_1)
print(parameter_three_Ar_SL_range_2)
print(parameter_three_Ar_SL_range_3)
```

## 2-1. When the spacing between the pore diameters was 0.5 angstrom

### 2-1-1. Compute the relative errors between each estimated value and experimental data

```python
pore_range_5step = []
Ar_SL_range_5step = []
Ar_BET_range_5step = []
AV_range_5step = []

for rrrrree in np.arange(3.9,7.9,0.5):
    
    Ar_SL_value_5step = linear_Ar_SL_three(rrrrree)
    Ar_BET_value_5step = linear_BET_three(rrrrree)
    AV_value_5step = linear_AV_three(rrrrree)
    
    pore_range_5step.append(rrrrree)
    
    Ar_SL_range_5step.append(Ar_SL_value_5step)
    Ar_BET_range_5step.append(Ar_BET_value_5step)
    AV_range_5step.append(AV_value_5step)

Ar_SL_error_DAM_5step_list = []
Ar_SL_error_DABA_5step_list = []
Ar_SL_error_BPDA_5step_list = []

Ar_BET_error_DAM_5step_list = []
Ar_BET_error_DABA_5step_list = []
Ar_BET_error_BPDA_5step_list = []

AV_error_DAM_5step_list = []
AV_error_DABA_5step_list = []
AV_error_BPDA_5step_list = []

for mmmeeee in Ar_SL_range_5step:
    
    Ar_SL_error_DAM_5step = np.abs((mmmeeee-Ar_SL_exp[0])/Ar_SL_exp[0])*100
    Ar_SL_error_DAM_5step_list.append(Ar_SL_error_DAM_5step)
    
    Ar_SL_error_DABA_5step = np.abs((mmmeeee-Ar_SL_exp[1])/Ar_SL_exp[1])*100
    Ar_SL_error_DABA_5step_list.append(Ar_SL_error_DABA_5step)
    
    Ar_SL_error_BPDA_5step = np.abs((mmmeeee-Ar_SL_exp[2])/Ar_SL_exp[2])*100
    Ar_SL_error_BPDA_5step_list.append(Ar_SL_error_BPDA_5step)
    
for mmeeeee in Ar_BET_range_5step:
    
    Ar_BET_error_DAM_5step = np.abs((mmeeeee-Ar_BET_exp[0])/Ar_BET_exp[0])*100
    Ar_BET_error_DAM_5step_list.append(Ar_BET_error_DAM_5step)
    
    Ar_BET_error_DABA_5step = np.abs((mmeeeee-Ar_BET_exp[1])/Ar_BET_exp[1])*100
    Ar_BET_error_DABA_5step_list.append(Ar_BET_error_DABA_5step)
    
    Ar_BET_error_BPDA_5step = np.abs((mmeeeee-Ar_BET_exp[2])/Ar_BET_exp[2])*100
    Ar_BET_error_BPDA_5step_list.append(Ar_BET_error_BPDA_5step)

for taxeee in AV_range_5step:
    
    AV_error_DAM_5step = np.abs((taxeee-AV_exp[0])/AV_exp[0])*100
    AV_error_DAM_5step_list.append(AV_error_DAM_5step)
    
    AV_error_DABA_5step = np.abs((taxeee-AV_exp[1])/AV_exp[1])*100
    AV_error_DABA_5step_list.append(AV_error_DABA_5step)
    
    AV_error_BPDA_5step = np.abs((taxeee-AV_exp[2])/AV_exp[2])*100
    AV_error_BPDA_5step_list.append(AV_error_BPDA_5step)
```

### 2-1-2. Estimate pore size distribution for DAM sample

```python
# create a LP minimization problem
Lp_prob_DAM_5step = p.LpProblem('Problem', p.LpMinimize)

# create problem variables
x1 = p.LpVariable("x1", lowBound = 0)
x2 = p.LpVariable("x2", lowBound = 0)
x3 = p.LpVariable("x3", lowBound = 0)
x4 = p.LpVariable("x4", lowBound = 0)
x5 = p.LpVariable("x5", lowBound = 0)
x6 = p.LpVariable("x6", lowBound = 0)
x7 = p.LpVariable("x7", lowBound = 0)

variable_list_DAM_5step=[x1,x2,x3,x4,x5,x6,x7,
                         1-x1-x2-x3-x4-x5-x6-x7]

Sum_AV_SL_BET_error_DAM_5step=0
Sum_SL_DAM_5step=0
Sum_BET_DAM_5step=0
Sum_AV_DAM_5step=0

for nnnnnne in range(len(variable_list_DAM_5step)):
    
    aaaaaae=(Ar_SL_error_DAM_5step_list[nnnnnne]+Ar_BET_error_DAM_5step_list[nnnnnne]+AV_error_DAM_5step_list[nnnnnne])*variable_list_DAM_5step[nnnnnne]
    Sum_AV_SL_BET_error_DAM_5step=Sum_AV_SL_BET_error_DAM_5step+aaaaaae

    bbbbbbe = Ar_SL_range_5step[nnnnnne]*variable_list_DAM_5step[nnnnnne]
    Sum_SL_DAM_5step=Sum_SL_DAM_5step+bbbbbbe
    
    cccccce = Ar_BET_range_5step[nnnnnne]*variable_list_DAM_5step[nnnnnne]
    Sum_BET_DAM_5step = Sum_BET_DAM_5step+cccccce
    
    cccchoco = AV_range_5step[nnnnnne]*variable_list_DAM_5step[nnnnnne]
    Sum_AV_DAM_5step = Sum_AV_DAM_5step + cccchoco

print(Sum_AV_SL_BET_error_DAM_5step) 

# objective function 
Lp_prob_DAM_5step += Sum_AV_SL_BET_error_DAM_5step

# mixing constraints
Lp_prob_DAM_5step += (1-x1-x2-x3-x4-x5-x6-x7) >= 0
Lp_prob_DAM_5step += Sum_BET_DAM_5step >= Ar_BET_exp[0]*0.92
Lp_prob_DAM_5step += Sum_BET_DAM_5step <= Ar_BET_exp[0]*1.08
Lp_prob_DAM_5step += Sum_AV_DAM_5step >= AV_exp[0]*0.92
Lp_prob_DAM_5step += Sum_AV_DAM_5step <= AV_exp[0]*1.08
Lp_prob_DAM_5step += Sum_SL_DAM_5step >= Ar_SL_exp[0]*0.92
Lp_prob_DAM_5step += Sum_SL_DAM_5step <= Ar_SL_exp[0]*1.08

# display the problem
print(Lp_prob_DAM_5step)

status_DAM_5step = Lp_prob_DAM_5step.solve()
print(p.LpStatus[status_DAM_5step])

final_variable_list_DAM_5step =[p.value(x1),p.value(x2),p.value(x3),p.value(x4),p.value(x5),p.value(x6),p.value(x7),
                                p.value(1-x1-x2-x3-x4-x5-x6-x7)]

fig=plt. figure(figsize =(13,10)) 
ax1613=fig.add_subplot(1,1,1)
plt. xlim(3.5,7.5)
plt. ylim(0,1.1)
plt. xticks(fontsize=28)
plt. tick_params(axis='both', width=4, length=10)
plt. yticks(fontsize=28)
for axis in ['top','bottom','left','right']:
    ax1613.spines[axis].set_linewidth(4)

plt.bar(pore_range_5step, final_variable_list_DAM_5step, color='blue', alpha=0.5, width=0.1, label='DAM')

plt.xlabel('Pore Diameter (Å)', fontsize=36)
plt.ylabel('Normalized Frequency (-)', fontsize=36)
plt.legend(frameon=True, loc = 'best',fontsize=24)
plt. savefig("./DAM_three_all_6_0420_05step.png",format='png')
plt.show()

for ooooooe in range(len(final_variable_list_DAM_5step)):
    print(pore_range_5step[ooooooe],final_variable_list_DAM_5step[ooooooe])

sum_Ar_SL_final_DAM_5step=0
sum_Ar_BET_final_DAM_5step=0
sum_AV_final_DAM_5step=0

for kkkkkke in range(len(final_variable_list_DAM_5step)):
    dddcdde = Ar_SL_range_5step[kkkkkke]*final_variable_list_DAM_5step[kkkkkke]
    sum_Ar_SL_final_DAM_5step=sum_Ar_SL_final_DAM_5step+dddcdde

    ddddragon = Ar_BET_range_5step[kkkkkke]*final_variable_list_DAM_5step[kkkkkke]
    sum_Ar_BET_final_DAM_5step=sum_Ar_BET_final_DAM_5step+ddddragon
    
    ssssupport = AV_range_5step[kkkkkke]*final_variable_list_DAM_5step[kkkkkke]
    sum_AV_final_DAM_5step=sum_AV_final_DAM_5step+ssssupport

ddddddee=np.abs((sum_Ar_SL_final_DAM_5step-Ar_SL_exp[0])/Ar_SL_exp[0])*100  
aaaapple=np.abs((sum_Ar_BET_final_DAM_5step-Ar_BET_exp[0])/Ar_BET_exp[0])*100
bbbbanana=np.abs((sum_AV_final_DAM_5step-AV_exp[0])/AV_exp[0])*100

print("Ar SL for DAM = "+str(sum_Ar_SL_final_DAM_5step))
print("error (%) ="+str(ddddddee))
print("Ar BET for DAM = "+str(sum_Ar_BET_final_DAM_5step))
print("error (%)"+str(aaaapple))
print("AV for DAM = "+str(sum_AV_final_DAM_5step))
print("error (%)"+str(bbbbanana))
```
